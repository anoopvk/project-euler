```python
primes=[2]
```

```python
import time
import math
start=time.time()
sum=2
for i in range(3,2000000):
    flag=1
    sqroot=math.sqrt(i)
    for j in primes:
        if j>sqroot:
            break
        if i%j==0:
            flag=0
    if flag==1:
        sum=sum+i
        print(sum)

end=time.time()
print(end-start)
```
